```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the cleaned and synthetic data
real_data = pd.read_csv("cleaned_diabetes_v2.csv")
synthetic_data = pd.read_csv("synthetic_diabetes_v2.csv")
```

```python
#Statistical Summary of the real data
real_data.describe()
```

```python
#Statistical summary of the synthetic data
synthetic_data.describe()
```

```python
#Calculating the average value Glucose, BMI & Age
print("Real Data Mean:")
print(real_data[['Glucose', 'BMI', 'Age']].mean())

print("\nSynthetic Data Mean:")
print(synthetic_data[['Glucose', 'BMI', 'Age']].mean())
```

```python
#Calculating the std value Glucose, BMI & Age
print("Real Data Standard Deviation:")
print(real_data[['Glucose', 'BMI', 'Age']].std())

print("\nSynthetic Data Standard Deviation:")
print(synthetic_data[['Glucose', 'BMI', 'Age']].std())
```

```python
#Histogram for Glucose
plt.figure(figsize=(8, 5))
sns.histplot(real_data['Glucose'], color='blue', label='Real', kde=True)
sns.histplot(synthetic_data['Glucose'], color='red', label='Synthetic', kde=True)
plt.legend()
plt.title('Glucose Distribution: Real vs Synthetic')
plt.show()
```

```python
#BMI
plt.figure(figsize=(8, 5))
sns.histplot(real_data['BMI'], color='blue', label='Real', kde=True)
sns.histplot(synthetic_data['BMI'], color='red', label='Synthetic', kde=True)
plt.legend()
plt.title('BMI Distribution: Real vs Synthetic')
plt.show()
```

```python
#age
plt.figure(figsize=(8, 5))
sns.histplot(real_data['Age'], color='blue', label='Real', kde=True)
sns.histplot(synthetic_data['Age'], color='red', label='Synthetic', kde=True)
plt.legend()
plt.title('Age Distribution: Real vs Synthetic')
plt.show()
```

```python
#Correlation heatmap- Real data
plt.figure(figsize=(10, 8))
sns.heatmap(real_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap - Real Data')
plt.show()
```

```python
#Correlation Heatmap- Synthetic data
plt.figure(figsize=(10, 8))
sns.heatmap(synthetic_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap - Synthetic Data')
plt.show()
```

```python
real_corr = real_data.corr()
synthetic_corr = synthetic_data.corr()

print("Real Data Correlation with Outcome:")
print(real_corr['Outcome'].sort_values(ascending=False))

print("\nSynthetic Data Correlation with Outcome:")
print(synthetic_corr['Outcome'].sort_values(ascending=False))
```
