```python
#imports
import pandas as pd
from sdv.single_table import CTGANSynthesizer
from sdv.metadata import SingleTableMetadata
```

```python
#loading the cleaned file
df=pd.read_csv("cleaned_diabetes_v2.csv")
```

```python
#Metadata for CTGAN training
metadata = SingleTableMetadata()
metadata.detect_from_dataframe(data=df)
```

```python
#code for saving the metedata for future purposes
metadata.save_to_json("metadata_v2.json")
```

```python
#Creating the CTGAN model
synthesizer = CTGANSynthesizer(
    metadata,
    epochs=1000,
    batch_size=100
)
```

```python
synthesizer.fit(df)
```

```python
#Genereation of synthetic rows
synthetic_data = synthesizer.sample(num_rows=len(df))
```

```python
#Displaying the first 5 rows
synthetic_data.head()
```

```python
#Saving the synthetic dataset
synthetic_data.to_csv("synthetic_diabetes_v2.csv", index=False)
```
